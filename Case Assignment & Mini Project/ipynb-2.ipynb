```python
# Count of Total players
player_count = len(purchase_data["SN"].value_counts())

#Which of players are females
female_players = purchase_data.loc[purchase_data["Gender"] == "Female"]
#Count of Females
count_of_female_players = len(female_players["SN"].value_counts())
#Percent of Females
percent_female = (count_of_female_players / player_count) * 100

#Which of players are male
male_players = purchase_data.loc[purchase_data["Gender"] == "Male"]
#Count of males
count_of_male_players = len(male_players["SN"].value_counts())
#Percent of Males
percent_male = (count_of_male_players / player_count) * 100

#Which of players are other
other_players = purchase_data.loc[purchase_data["Gender"] == "Other / Non-Disclosed"]
#Count of other
count_of_other_players = len(other_players["SN"].value_counts())
#Percent of other
percent_other = (count_of_other_players / player_count) * 100

# Summary Data Frame
summary_gender_percent_df = pd.DataFrame([{
    "Gender": "Female", "Total Count": count_of_male_players, 
    "Percentage of Players": percent_female}, 
    {"Gender": "Male", "Total Count": count_of_female_players, 
     "Percentage of Players": percent_male}, 
    {"Gender": "Other / Non-Disclosed", "Total Count": count_of_other_players, 
     "Percentage of Players": percent_other
    }], 
columns=["Gender", "Total Count", "Percentage of Players"])
summary_gender_percent_df["Percentage of Players"] = summary_gender_percent_df["Percentage of Players"].map("{0:.2f}%".format)

summary_gender_percent_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Establish Bins for Ages 
bins = [0, 9, 14, 19, 24, 29, 34, 39, 50]
bin_ranges = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

# Categorize the existing players using the age bins. Hint: use pd.cut and make new column
purchase_data["Age Group"] = pd.cut(purchase_data["Age"], bins=bins, labels=bin_ranges)
purchase_data

# Create a GroupBy Object Based Upon "Age Group"
age_group = purchase_data.groupby("Age Group")

# Count Total Players by Age Category
total_count_age = age_group["SN"].nunique()

# Calculate Percentages by Age Category 
percentage_by_age = round(total_count_age / player_count * 100,2)

#Summary DataFrame
age_demographics_df = pd.DataFrame({
    "Total Count": total_count_age, 
    "Percentage of Players": percentage_by_age
})

age_demographics_df["Percentage of Players"] = age_demographics_df["Percentage of Players"].map("{0:.2f}%".format)

age_demographics_df
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Identify the Most Popular Items
list_popular_items = purchase_data[["Item ID", "Item Name", "Price"]]

# GroupBy "Item ID" & Item Name
popular_items = list_popular_items.groupby(["Item ID","Item Name"])

# Calculate Purchase Count
item_purchases_count = popular_items["Price"].count()

# Calculate Item Price
item_price = popular_items["Price"].sum()

# Calculate Total Purchase Value 
item_purchase_value = item_price / item_purchases_count

# Create Summary DataFrame
most_popular_items_df = pd.DataFrame({
    "Purchase Count": item_purchases_count, 
    "Item Price": item_purchase_value,
    "Total Purchase Value": item_price
})

sorted_most_popular_items = most_popular_items_df.sort_values(["Purchase Count"], ascending=False).head()
most_popular_items_df["Item Price"] = most_popular_items_df["Item Price"].map("${0:.2f}".format)
most_popular_items_df["Total Purchase Value"] = most_popular_items_df["Total Purchase Value"].map("${0:.2f}".format)

sorted_most_popular_items
```
